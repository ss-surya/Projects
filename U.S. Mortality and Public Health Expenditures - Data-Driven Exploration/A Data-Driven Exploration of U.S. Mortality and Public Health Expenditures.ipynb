## Data loading from web API

```python
# Importing Life expectency and Mortality data from web API
url = "https://data.cdc.gov/resource/w9j2-ggv5.json"
       
response = requests.get(url)
lm_data = json.loads(response.content)
lm_df = pd.DataFrame()
# Iterate over the dictionaries in the JSON object
for i in lm_data:
    # Extract the data from the dictionary
    dict = {"Year" : [i['year']],
            "Race" : [i['race']],
            "Sex" : [i['sex']],
            "Average Life Expectancy" : i.get('average_life_expectancy', None),
            "Mortality" : [i['mortality']]}
    # Create a temporary DataFrame for each dictionary
    temp_df = pd.DataFrame(dict)
    # Append the temporary DataFrame to the main DataFrame
    lm_df = pd.concat([lm_df, temp_df],ignore_index = True)

# Used astype() to change the datatype
lm_df['Year'] = lm_df['Year'].astype(int)
lm_df['Average Life Expectancy'] = lm_df['Average Life Expectancy'].astype(float)
lm_df['Mortality'] = lm_df['Mortality'].astype(float)
    
lm_df.head()  
```

## Class Structure for performing EDA

The use of a class in this code is justified by encapsulating related functionalities and data within a modular and organized structure. 

The class named __eda__ serves as a container for functions that collectively perform exploratory data analysis (EDA) tasks. By grouping these functions together, the code becomes more __readable, maintainable, and reusable__. The class abstraction allows for a clear separation of concerns, enabling users to interact with distinct EDA methods without being concerned about their internal implementations. Additionally, using a class promotes __code reusability__; different instances of the eda class can be created and utilized for EDA on various datasets or with different configurations. This approach adheres to the principles of object-oriented programming, facilitating a modular and extensible design that enhances code organization and scalability.

Overall, the use of a class in this context contributes to a more structured, modular, and maintainable codebase, facilitating effective exploratory data analysis.

```python
class eda:
    '''Class containing functions to perform Exploratory Data Analysis
    Functions:
    Univariate_EDA: Performs univariate exploratory data analysis on a specified column, displays summary statistics and appropriate plots.
    Bivariate_EDA: Performs bivariate exploratory data analysis on two specified columns, displays appropriate plots.
    Multivariate_EDA: Performs bivariate exploratory data analysis on multiple specified columns, displays appropriate plots.
    
    Refer to docstrings of functions for specific details    
    ''' 
    def Univariate_EDA(self, df, col):    
        """Performs univariate exploratory data analysis on a specified column, displaying summary statistics, unique values for categorical columns, and visualizations such as bar plots, histograms, box plots.        
        Args:
        df: A Dataframe to perform Univariate analysis
        col: Column in the df to perform Univariate analysis
        Returns:
        None
        """
        # get summary statistics
        print('Summary Statistics')
        display(df[col].describe())

        #Categorical Variable
        if df[col].dtype == 'object':
            
            # display unique values of Gender
            print('Unique Values')
            display(df[col].unique())

            # create a bar plot of the value counts   
            plt.figure(figsize=(4,4))
            plt.title(col)

            # get value counts for the variable
            frequency = df[col].value_counts()
            sns.barplot(x=frequency.index, y=frequency.values, palette="viridis")
            plt.title(col)
            plt.ylabel("Counts")
            plt.tick_params(axis='x', rotation=90)

        # Numerical Variable
        elif df[col].dtype == 'float64' or  df[col].dtype == 'int32' or  df[col].dtype == 'int64': 

            plt.figure(figsize=(8,4))
            plt.suptitle(col)
            
            # create a histogram plot 
            plt.subplot(1,2,1)
            sns.histplot(data=df, x=col, kde=True, color='skyblue')
            
            # create a boxplot 
            plt.subplot(1,2,2)
            sns.boxplot(data= df, y= col, color='green')  

        # Display plot
        plt.tight_layout()
        plt.show()    

        return None 
    
    
    def Bivariate_EDA(self, df, col1, col2):
        """Conducts bivariate exploratory data analysis by generating visualizations based on two columns, creating violin plots for categorical vs numeric or scatter plots for numeric vs numeric analysis.        
        Args:
        df: A Dataframe to perform Bivariate analysis
        col1: Column in the df to perform Bivariate analysis
        col2: Column in the df to perform Bivariate analysis
        Returns:
        None
        """

        # Plotting a violin plot for categorical vs numeric analysis
        if df[col1].dtype == 'float' and df[col2].dtype == 'object': 
            plt.figure(figsize=(5,4))            
            sns.violinplot(x=col2, y=col1, data=df) 

        elif df[col1].dtype == 'object' and df[col2].dtype == 'float': 
            plt.figure(figsize=(5,4))
            sns.violinplot(x=col1, y=col2, data=df)
            
        # Plotting a scatter plot for numeric vs numeric analysis
        elif (df[col1].dtype == 'int32' or df[col1].dtype == 'int64' or df[col1].dtype == 'float64') and (df[col2].dtype == 'int32' or df[col2].dtype == 'int64' or df[col2].dtype == 'float64'): 
            plt.figure(figsize=(5,4))
            if col1=='Year':
                sns.lineplot(x=col1, y=col2, data=df)
            else:
                sns.scatterplot(x=col1, y=col2, data=df, alpha=0.5)    

        # Set plot title and rotate x-axis labels for better visibility
        plt.title(col1 + ' vs ' + col2)
        plt.tick_params(axis='x', rotation=90)

        # Display plot
        plt.tight_layout()
        plt.show()   
        return None  
    
    def Multivariate_EDA(self, df, col1, col2, col3, col4, type_of_graph):
        """Explores relationships among three variables by creating a scatter plot matrix, displaying 
        scatter plots with color-coding based on a third categorical variable.
        
        Args:
        df: A Dataframe to perform Multivariate analysis
        col1: Column in the df to perform Multivariate analysis
        col2: Column in the df to perform Multivariate analysis
        col3: Column in the df to perform Multivariate analysis
        col4: Column in the df to perform Multivariate analysis
        Returns:
        None"""
        
        # Create a figure with two subplots side by side
        plt.figure(figsize=(5,4))
        
        if type_of_graph == "SCATTER":            
            sns.scatterplot(x=col1, y=col2, hue = col3 , data=df, alpha=0.3)
            plt.title(col1 + ' vs ' + col2  + ' vs ' + col3)
            plt.legend(title = col3, bbox_to_anchor=(1, 1))
        
        elif type_of_graph == "LINE":
            temp_df = lm_df[(lm_df[col3]!='All Races') & (lm_df[col1]!='Both Sexes')]            
            sns.lineplot(x=col4, y=col2, hue=temp_df[col1] + '_' + temp_df[col3], data=temp_df)
            
        # Display plot
        plt.tight_layout()
        plt.show() 

ua = eda()
ba = eda()
ma = eda()
```

# DATA PREPARATION

## 1) Handling Missing data values

In the dataset, data for the 'White' Race and 'Male' gender is absent after the Year 1947. To address this gap, we propose to estimate the life expectancy and Mortality rate for males by utilizing the combined values of both genders and subtracting the corresponding female values. This approach aims to fill in the missing information for 'White' males post-1947, providing a more comprehensive and complete dataset for further analysis.

```python
# Filter the DataFrame for White Race and post-1947 Years, drop the 'Race' column
filtered_df = lm_df[(lm_df['Year'] >= 1948) & (lm_df['Race'] == 'White')].reset_index(drop=True).drop('Race', axis=1)
# Pivot the filtered DataFrame
reshaped_Df = filtered_df.pivot(index=['Year'], columns='Sex', values=['Average Life Expectancy', 'Mortality'])

# Calculate new columns for Male based on existing columns
reshaped_Df.loc[:, ('Average Life Expectancy', 'Male')] = (2 * reshaped_Df[('Average Life Expectancy', 'Both Sexes')]) - reshaped_Df[('Average Life Expectancy', 'Female')]
reshaped_Df.loc[:, ('Mortality', 'Male')] = (2 * reshaped_Df[('Mortality', 'Both Sexes')]) - reshaped_Df[('Mortality', 'Female')]
# Stack the pivoted DataFrame and reset the index
reshaped_Df = reshaped_Df.stack().reset_index()
# Add 'Race' column with value 'White'
reshaped_Df['Race'] = 'White'
# Filter rows where 'Sex' is 'Male'
reshaped_Df = reshaped_Df[reshaped_Df['Sex'] == 'Male']

# Concatenate the original and reshaped DataFrames, sort, and reset the index
lm_df_new = pd.concat([lm_df, reshaped_Df]).sort_values(by=['Year', 'Race', 'Sex']).reset_index(drop=True)

# Display the resulting DataFrame
lm_df_new
```

## 2) Merging Data

Combine both datasets using the "Year" column to create a merged dataframe that includes information on average life expectancy, Mortality rate, and total national health expenditure in constant 2021 dollars for all, black, and white Races, encompassing both male and female Sexes for the Years 1970 to 2018.

```python
merged_df = pd.merge(lm_df_new, he_data, on='Year')

merged_df.head()
```

Fill null values in the 'life_expectancy' column with values from the previous Years within the same 'Sex' and 'Race' category. Since the data represents a time series with a gradual upward trend, this approach ensures a continuous and coherent progression of life expectancy for individuals with matching characteristics.

```python
# Sort the DataFrame by 'Sex', 'Race', 'Year' 
merged_df = merged_df.sort_values(by=['Sex', 'Race', 'Year'])

# Fill null values in 'Average Life Expectancy' with values from previous Years within the same 'Sex' and 'Race'
merged_df['Average Life Expectancy'] = merged_df.groupby(['Sex', 'Race'])['Average Life Expectancy'].transform(lambda x: x.ffill())
```

## Investigative Analysis 3

To employ predictive modeling techniques for forecasting the future Mortality rate in the United States. This involves leveraging historical data, statistical methods, and possibly machine learning algorithms to project Mortality trends. The goal is to provide insights into the likely trajectory of Mortality rates, enabling better preparation and informed decision-making in healthcare planning.

#### Step 1: Prepare the data

```python
# Preparing dataset
model_df = merged_df[(merged_df['Sex'] == 'Both Sexes') & (merged_df['Race'] == 'All Races')]
model_df = model_df[['Year', 'Total national health expenditures']].set_index('Year')
model_df = model_df.rename(columns={'Total national health expenditures': 'Expense'})
model_df.info()
```

#### Step 2: Visualize the data

```python
# plotting the original data
model_df.plot(title='National health expenditures of US over Years')
```

#### Step 3: Fitting the data to Model

```python
model_df['Expense First Difference'] = model_df['Expense'] - model_df['Expense'].shift(1)
model_df.head()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(model_df['Expense First Difference'].dropna(),lags=23,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(model_df['Expense First Difference'].dropna(),lags=23,ax=ax2)
```

```python
# Split the data to train and test
train = model_df[:45]
test = model_df[45 :]

# Fit the data to ARIMA model
#p=1, d=1, q=15
model = sm.tsa.arima.ARIMA(model_df['Expense'],order=(1,2,15))
model_fit = model.fit()

# Compare Actual an Predicted Values
model_df['Predicted Expense']=model_fit.predict(start=45,end=49,dynamic=True)
model_df[['Expense','Predicted Expense']].plot(figsize=(5,5))
```

#### Step 4: Model Evaluation

```python
model_eval = model_df[['Expense', 'Predicted Expense']].tail(4)

# Mean Absolute Error (MAE)
mae = mean_absolute_error(model_eval["Expense"], model_eval["Predicted Expense"])
print(f"Mean Absolute Error (MAE): {mae:.4f}")

# Mean Squared Error (MSE)
mse = mean_squared_error(model_eval["Expense"], model_eval["Predicted Expense"])
print(f"Mean Squared Error (MSE): {mse:.4f}")

# R-squared (Coefficient of determination)
r2  = r2_score(model_eval["Expense"], model_eval["Predicted Expense"])
print(f"R-squared: {r2:.4f}")
```

- Lower MSE and MAE indicate better model fit.
- Higher R-squared indicates a stronger linear relationship between predicted and actual values.

Super high R-squared values might suggest over-fitting. Future extensions might be needed to optimize the model.

#### Step 5: Predicting Future Values

```python
# Forecast future values
forecast_steps = 5  # Number of steps to forecast into the future
future_forecast = model_fit.forecast(steps=forecast_steps)

future_forecast = future_forecast.to_frame()
#future_forecast.rename(columns={'predicted_mean': 'Forecast Expense'}, inplace=True)
future_forecast.rename(columns={'predicted_mean': 'Expense'}, inplace=True)
future_forecast
```

```python
#Plot the data along with forecast

model_df['flag'] = 'Historical'
future_forecast['flag'] = 'Forecast'

# calculation to help to connect the gap in the graph
helper = future_forecast[['Expense']].head(1)
helper['flag'] = 'Historical'

plot_df = pd.concat([model_df[['Expense','flag']], helper, future_forecast])
plot_df

sns.lineplot(x=plot_df.index, y=plot_df['Expense'], hue = plot_df['flag'])
plt.title("National Health Expenditure Forecast", fontsize=16)
```

 The plot unmistakably illustrates a increasing trend in the future National Health Expenditures, indicating positive conditions in the healthcare situation in the USA.
